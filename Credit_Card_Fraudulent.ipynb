This dataset is highly unbalanced

```python
# 0 -- represents nomral transaction
# 1 -- represents fraudulent transaction
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

Under Sampling

build a sample dataset containig similar distribution of normal transactions and fraudulent transactions

```python
# number of fraudulent transactions are --> 492
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample , fraud] , axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Spliting the data into feature and targets

```python
X = new_dataset.drop(columns = 'Class' , axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y , test_size=0.2 , stratify=Y,random_state=2 )
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
# trainig the logistic regression model with trainig data
model.fit(X_train , Y_train)
```

Model Evaluation

Accuracy Scoore

```python
# accuracy on trainig data
X_train_prediciton = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediciton , Y_train)
```

```python
print('Accuracy on the trainig data :',training_data_accuracy )
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print(" Accuracy score on test data " , test_data_accuracy)
```
